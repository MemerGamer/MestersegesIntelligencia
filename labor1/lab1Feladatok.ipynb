```python
# library
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# 4. Logisztikus vagy szigmoid alakú karakterisztika
def SigmoidBinary(a, t, theta):
    return 1 / (1 + np.exp(-a * t - theta))


def tangensHyperbolic(a, t, theta):
    return (1 - np.exp(-a * t - theta)) / (1 + np.exp(-a * t - theta))
```

II. 1. Programozzunk le az elmélet alapján egy neuront. A súlytényezőket inicializáljuk véletlenszerűen.
egyszeru neuron.png

```python
# 1 réteg felépítése
import random

# bemenetek
x = []
# Set a length of the list to 3
for i in range(0, 3):
    # any random float between -1 to 1
    temp = round(random.uniform(0, 1), 2)
    x.append(temp)

print(x)

# sulyok
w = []
# Set a length of the list to 3
for i in range(0, 3):
    # any random float between -1 to 1
    temp = round(random.uniform(-1, 1), 2)
    w.append(temp)

print(w)

# bias
teta = 0

# segedvaltozo
inger = 0.0
for i in range(0, 3):
    inger += (w[i] * x[i]) - teta
    print(round(inger, 2))


# a kovetkezo aktivacios fuggvenyt fogjuk használni: lépcsõfüggvény
def stairfunction(t, theta):
    return (t <= theta) * -1 + (t >= theta) * 1


y = stairfunction(inger, teta)
print(f"A lépcsõfüggvény eredménye: {y}")
```

I.3.   Ábrázoljuk az "a" és "teta" paraméterek függvényében egy tetszőlegesen választott aktivációs függvényt(minimum 2 akt. fg.)


```python
# I - lineáris
def linearActivationFunction(a, t, theta):
    return a * t - theta


a = 1
t = np.linspace(-5, 5)
theta = np.linspace(0.5, 1.5, num=3)
szin = [". ", "--", "-."]

for i in range(theta.size):
    y = linearActivationFunction(a, t, theta[i])
    plt.plot(t, y, szin[i], label=f"theta = {theta[i]}", linewidth=2)
    plt.legend()

plt.title('Lineáris akt. fg. az "theta" paraméter függvényében')
plt.show()
```

```python
# II - Gauss
def gauss(a, c, t, sigma):
    return np.exp((-np.square(t * a - c)) / 2 * (sigma * sigma))


a = 1
c = 1
t = np.linspace(-5, 5)

theta = np.linspace(0.5, 1.5, num=3)
szin = [". ", "--", "-."]

for i in range(theta.size):
    y = gauss(a, c, t, theta[i])
    plt.plot(t, y, szin[i], label=f"theta = {theta[i]}", linewidth=2)
    plt.legend()

plt.title('Gauss akt. fg. az "theta" paraméter függvényében')
plt.show()
```

I.4.   Ábrázoljuk grafikusan a szigmoid és tangens hiperbolikus függvények deriváltját.



```python
def sigmoidDerivative(a, t, theta):
    return a * np.exp(a * t + theta) / np.square(1 + np.exp(a * t + theta))

a = 1
t = np.linspace(-5, 5)
theta = np.linspace(0.5, 1.5, num=3)
szin = [". ", "--", "-."]

for i in range(theta.size):
    y = sigmoidDerivative(a, t, theta[i])
    plt.plot(t, y, szin[i], label=f"theta = {theta[i]}", linewidth=2)
    plt.legend()
    
plt.title('Sigmoid akt. fg. deriváltja az "theta" paraméter függvényében')
```

```python
def tanhDerivative(a, t, theta):
    return 2*a*np.exp(a*t+theta) / np.square(1 + np.exp(a*t+theta))

a = 1
t = np.linspace(-5, 5)
theta = np.linspace(0.5, 1.5, num=3)
szin = [". ", "--", "-."]

for i in range(theta.size):
    y = tanhDerivative(a, t, theta[i])
    plt.plot(t, y, szin[i], label=f"theta = {theta[i]}", linewidth=2)
    plt.legend()
    
plt.title('Tg. akt. fg. deriváltja az "theta" paraméter függvényében')
```
