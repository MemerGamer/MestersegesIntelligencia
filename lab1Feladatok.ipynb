I. 1.
Ábrázoljuk grafikusan és tanulmányozzuk a szigmoid és tangens hiperbolikus függvényeket az "a" paraméter függvényében.

```python
#library
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#4. Logisztikus vagy szigmoid alakú karakterisztika
def SigmoidBinary(a, t, theta):
    return 1/(1+np.exp(-a*t-theta))

def tangensHyperbolic(a, t, theta):
    return (1-np.exp(-a*t-theta))/(1+np.exp(-a*t-theta))
```

```python
plt.plot([1,2,3], [1,2,3], 'go-', label='line 1', linewidth=2)
plt.plot([1,2,3], [1,4,9], 'rs',  label='line 2')
plt.axis([0, 4, 0, 10])
plt.legend()
```

```python
a=np.linspace(0.5, 1.5, num=3)
t = np.linspace(-5, 5)
theta = 0
szin=['. ','--','-.']

for i in range(a.size):
  y = tangensHyperbolic(a[i], t, theta)
  plt.plot(t, y, szin[i], label=f"a = {a[i]}", linewidth = 2)
  plt.legend()

plt.title('Tg. akt. fg. az "a" paraméter függvényében')
plt.show()
```

```python
a=np.linspace(0.5, 1.5, num=3)
t = np.linspace(-5, 5)
theta = 0
szin=['. ','--','-.']

for i in range(a.size):
  y = SigmoidBinary(a[i], t, theta)
  plt.plot(t, y, szin[i], label=f"a = {a[i]}", linewidth = 2)
  plt.legend()

plt.title('Sigmoid akt. fg. az "a" paraméter függvényében')
plt.show()
```

II. 1. Programozzunk le az elmélet alapján egy neuront. A súlytényezőket inicializáljuk véletlenszerűen.
egyszeru neuron.png

```python
#1 réteg felépítése
import random

#bemenetek
x= []
# Set a length of the list to 3
for i in range(0, 3):
    # any random float between -1 to 1
    temp = round(random.uniform(0, 1), 2)
    x.append(temp)

print(x)

#sulyok
w = []
# Set a length of the list to 3
for i in range(0, 3):
    # any random float between -1 to 1
    temp = round(random.uniform(-1, 1), 2)
    w.append(temp)

print(w)

#bias
teta = 0;

#segedvaltozo
inger = 0.0
for i in range(0, 3):
  inger += ((w[i]*x[i])-teta)
  print(round(inger, 2))

# a kovetkezo aktivacios fuggvenyt fogjuk használni: lépcsõfüggvény
def stairfunction(t, theta):
    return (t<=theta) * -1 + (t>=theta) * 1

y = stairfunction(inger, teta)
print(f"A lépcsõfüggvény eredménye: {y}")
```
